```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# reshape(-1,1)
# Linear Regression expects X to be a 2D array.
# Original shape: (5,)
# New shape: (5,1)

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
```

```python
y = np.array([2, 4, 6, 8, 10])
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y)

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Dataset")

plt.show()
```

```python
# Train Model

model = LinearRegression()

# fit() trains the model.
# It learns the relationship between X and Y.

model.fit(x, y)
```

```python
# coef_ stores the slope (m)

print("Slope:", model.coef_)

print("Intercept :", model.intercept_)
```

```python
# Predict the value of y when x = 6

prediction = model.predict([[6]])

print(prediction)
```

```python
plt.scatter(x, y)

plt.plot(
    x,
    model.predict(x)
)
```

```python
score = model.score(x, y)
print(score)
```

### # Notes

Linear Regression Equation:

y = mx + c

m = slope

c = intercept

Goal:
Find the best fitting line.
